# DISCRETE STOCHASTIC PROCESSES

## TUTORIAL

A discrete stochastic process is a countably indexed collection of random variables $(X_n)_{n\in \mathbb Z}$. A stochastic process differs conceptually from, e.g., an infinite dimensional random vector in that the random variables in the process occur sequentially in time rather than simultaneously.

In `pystochastica` discrete processes are built on top of the `DSP` class, found at `pystochastica.discrete.dsp.DSP`. In this tutorial we will demonstrate the discrete stochastic processes written in `pystochastica`.

### LIBRARY IMPORTS

```python
from pystochastica.discrete.dsp import RandWalk
```

### THE RANDOM WALK

the Random Walk process is a discrete stochastic process $(X_n)_n$ where $X_n$ is the total distance travelled after the $n$-th step. The process is seeded by an initial probability distribution, where the walker takes one step forward with probability $p$ and one step backward with probablity $1-p$. Evidently, the expected distance travelled is *zero*, however the volatility grows linearly with the number of steps.

The $n$-th variable $X_n$ is given by the $n$-fold sum, $X_n = \sum_{i=0}^n P_i$ where, for each $i$, $P_i$ is the random variable given by $+x$ with probability $p$ and $-x$ with probability $1-p$, where $x$ is the distance travelled in any given step. 

In ``pystochastica`` the Random Walk process is found at `pystochastica.discrete.dsp.RandWalk`. By default it is seeded by probability $p = 0.5$, however this can be changed as desired. We initialise `RandWalk` below with a custom probability and a number of total steps to take.

```python
rw = RandWalk(time_steps=50, p=0.65)
```

#### THE TOTAL WALK PLOT

The following is a plot of `RandWalk`, given by sampling the $n$-variable in the process. As ought to be evident from the plot, the volatiliy is growing as we go further along the process.


```python
rw.plt()
```

#### CUMULATIVE DISTANCE

The `plt_walk()` method plots the cumulative total distance after a variable number of steps. Below we see this cumulative total after `1000` steps. 

```python
rw.plt_walk(steps=1000)
```

#### CUMULATIVE DISTANCE SUBPLOTS

To see multiple, independent simulations of the cumulative distance plots, we can call the `plt_walks()` method. 

**Note.** *The number of plots (i.e., simulations) coincides with total number of time steps passed when the `RandWalk` was initialised.*


```python
rw2 = RandWalk(time_steps=8, p=0.65)
rw2.plt_walks(steps=1000, FIGSIZE=(10, 5))
```
